**VISUALIZATION**

```python
#Line plot on distribution of population in China over the decades
years = ["1970","1980","1990","2000","2010","2015","2020","2022"]
china_population = data.loc[41, [f"{year} Population" for year in years]].values
plt.plot(years, china_population, marker='*', label = "China", color='blue')

plt.title('Population Growth Over Time In China')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

***Inference:-***

  There is rapid increase in population in the 70's decades but it gradually slows down as years pass by. And there is slight decrease after 2020 (Reasons: Covid-19)

```python
years = ["1970","1980","1990","2000","2010","2015","2020","2022"]
rank = [1,2,3,4,5]
countries = list(data[data['Rank'].isin(rank)]['Country/Territory'])
print(countries)

colors = ['red','green','blue','orange','crimson']

plt.figure(figsize=(10, 6))

for country, color in zip(countries, colors):
    country_population = data.loc[data['Country/Territory'] == country, [f"{year} Population" for year in years]].values.flatten()
    plt.plot(years, country_population, marker='*', label=country, color=color)

plt.title('Population Growth Over Time of Countires ranked Top 5')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

***Inference:-***

India was far behind China in the 70's but now is on it's way to overtake China in terms of Population. Whereas there's been very slight increase in population of the next 3 most populous countries over the past 5 decades.  

```python
#Bar Chart on distribution of the countries in different continents
continent_data = pd.DataFrame(data['Continent'].value_counts())
print(continent_data)
```

```python
continents = list(continent_data.index)
print(continents)
countries = [continent_data['Continent'][i] for i in range(len(continent_data))]
print(countries)
```

```python
#Categorical variable distribution

plt.bar(continents, countries, color = "orange")
plt.title('Distribution of Countires in Continents')
plt.xlabel('Continents')
plt.ylabel('Number of Countries')
plt.xticks(rotation=30)
plt.show
```

```python
density = list(data.nlargest(10,'Density (per km²)')['Density (per km²)'])
print(density)

countries = list(data.nlargest(10,'Density (per km²)')['Country/Territory'])
print(countries)

plt.figure(figsize=(10, 6))
plt.bar(countries, density, color = 'red')
plt.title('10 Most Densely Populated Countries/Territories in the World')
plt.xlabel('Countires/Territories')
plt.ylabel('Density (per km²)')
plt.xticks(rotation=30)
plt.show
```

```python
european_countries = data[data['Continent'] == 'Europe']
countries = list(european_countries.nlargest(20,'Growth Rate')['Country/Territory'])
print(countries)

growth = list(european_countries.nlargest(20,'Growth Rate')['Growth Rate'])
print(growth)

plt.figure(figsize=(10, 6))
plt.bar(countries, growth, color = 'green')
plt.title('Top 20 Countries with highest Growth Rate in Europe')
plt.xlabel('Countires/Territories')
plt.ylabel('Population Growth Rate')
plt.xticks(rotation=90)
plt.show
```
